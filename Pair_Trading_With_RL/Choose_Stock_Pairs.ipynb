```python
import numpy as np
import pandas as pd
df_full = pd.read_csv('15stock.csv')
df_full.head()
```

```python
df_full = df_full.rename(columns={"Unnamed: 0": "date"})
df_full.head()
```

```python
df_new = pd.pivot(df_full, index="date", columns="windcode", values="close")
df_new.head()
```

```python
df_new = pd.DataFrame(df_new, columns = [ '000014.SZ', '000020.SZ'])
df_new = df_new.rename(columns={"000014.SZ": "stock1", "000020.SZ" : "stock2"})
df_new.head()
```

```python
df_new.count()
```

```python
df_new[["stock1"]].rename(columns={"stock1": "close"}).to_csv("stock1.csv")
df_new[["stock2"]].rename(columns={"stock2": "close"}).to_csv("stock2.csv")
```
